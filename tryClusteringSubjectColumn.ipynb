```python
subjectColumn = os.getcwd()+ "/result/subject_column/subject_column.csv"
f = open(subjectColumn, encoding='latin1', errors='ignore')
gt_CSV = pd.read_csv(f)
dict_columns = {}
table_features = {}
table_feature = "result/subject_column/test_subject_column.csv"
ft = open(table_feature, encoding='latin1', errors='ignore')
features_csv = pd.read_csv(ft)
```

```python
print(features_csv)
```

```python
list_features = []
for index, value in features_csv.iterrows():
    feature = np.fromstring(value[1].replace('\n', ' ').strip('[]'), dtype=float, sep=' ')
    feature = feature.reshape((1, feature.shape[0]))
    list_features.append(list(feature[0]))
features = np.array(list_features)
print(features)



```

```python
from sklearn.cluster import DBSCAN
from sklearn import datasets

# Load the iris dataset
X = features

print(X)

# Create an instance of DBSCAN
db = DBSCAN(eps=0.3, min_samples=10)

# Fit the model to the data
db.fit(X)

# Predict the cluster labels for each data point
labels = db.labels_
```

```python
from sklearn.cluster import DBSCAN
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {'eps': list(np.arange(start=0.1, stop=5, step=0.1)),
              'min_samples': list(np.arange(start=2, stop=25, step=1))}

# Create an instance of DBSCAN
db = DBSCAN()

# Create an instance of GridSearchCV
grid = GridSearchCV(db, param_grid, cv=5,scoring='accuracy')

# Fit the grid search to the data
grid.fit(X)

# Get the best parameters
best_params = grid.best_params_
print(best_params)
```
